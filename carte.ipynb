# Visualisation Géographique


La visualisation géographique permet de représenter des données sur une carte.

Elle nous permet de visualiser des localisations, des déplacements,
des regroupements spatiaux, ou des données démographiques.


### Bibliothèques utiles pour la visualisation géographique :
- `geopandas` : pour manipuler des données spatiales
- `folium` : pour créer des cartes interactives
- `matplotlib` : pour des cartes statiques

### Démonstration

```python
# Imports
import folium
import pandas as pd

# Chargement des données du dataset
df = pd.read_csv("./Dataset/Villes.csv")

# Création d'une carte
map = folium.Map(location=[46.5, 2.2], zoom_start=6)

# Ajout des villes
for _, row in df.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=f"<b>{row['ville']}</b><br>Population : {row['population']:,}",
    ).add_to(map)

# Enregistre la carte dans un fichier HTML
map.save("Carte_Villes.html")

# Affiche la carte
map
```
