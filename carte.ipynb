# Visualisation Géographique


La visualisation géographique permet de représenter des données sur une carte.

Elle nous permet de visualiser des localisations, des déplacements,
des regroupements spatiaux, ou des données démographiques.


### Bibliothèques utiles pour la visualisation géographique :
- `geopandas` : pour manipuler des données spatiales
- `folium` : pour créer des cartes interactives
- `matplotlib` : pour des cartes statiques

### Démonstration

```python
# Imports
import folium
import pandas as pd

# Chargement des données du dataset des villes de France
df_villes = pd.read_csv("./Dataset/Villes.csv")

# Chargement des données du dataset des capitales d'Europe
df_capitales = pd.read_csv("./Dataset/Capitales.csv")

# Création d'une carte
map = folium.Map(location=[46.5, 10], zoom_start=4)

# Ajout des villes
for _, row in df_villes.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=f"<b>{row['ville']}</b><br>Population : {row['population']:,}",
        icon=folium.Icon(color='red')
    ).add_to(map)


# Ajout des capitales
for _, row in df_capitales.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=f"<b>{row['capitale']}</b><br>Pays : {row['pays']}",
        icon=folium.Icon(color='blue', icon='flag')
    ).add_to(map)

# Enregistre la carte dans un fichier HTML
map.save("Carte_Europe.html")

# Affiche la carte
map
```
